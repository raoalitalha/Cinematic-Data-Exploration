## This column is highly subjective in my point of view. So i am taking the following action for futture but not removing outliers and doing anything else.

- Applying the log transformation of log1p on the colum and creating a new colum `popularity_log`
- Secondly , i am applying `boxcox` transformation and creating a new colum `popularity_boxcox`.


#### All the above transformations is for future i=usage , as to my understanding from the metadata the 'popularity ' column is subjective.

```python
# Apply log transformation to the column 'popularity'
df['popularity_log'] = np.log1p(df['popularity'])
```

```python
df.head(2)
```

```python
sns.boxplot(df['popularity_log'])
```

```python
# Create a histogram of the column 'popularity_log'
sns.histplot(df['popularity_log'])
```

```python
df['popularity_log'].describe()
```

```python

from scipy import stats

# 'popularity' is the column to transform
popularity_data = df['popularity'].values  # Extract the 'popularity' column as a NumPy array

# Estimate the optimal lambda (λ) value
optimal_lambda = stats.boxcox_normmax(popularity_data)

# Apply the Box-Cox transformation
transformed_popularity = stats.boxcox(popularity_data, lmbda=optimal_lambda)

# Create a new column 'popularity_boxcox' to store the transformed values
df['popularity_boxcox'] = transformed_popularity
```

```python
df.head(2)
```

```python
sns.boxplot(df['popularity_boxcox'])
```

```python
sns.histplot(df['popularity_boxcox'])
```

```python
def winsorize(data, lower_percentile=1, upper_percentile=99):
    lower_limit = np.percentile(data, lower_percentile)
    upper_limit = np.percentile(data, upper_percentile)
    data = np.where(data < lower_limit, lower_limit, data)
    data = np.where(data > upper_limit, upper_limit, data)
    return data
```

```python
lower_percentile = 1  # Example: 1st percentile
upper_percentile = 99  # Example: 99th percentile

df['popularity_winsorized'] = winsorize(df['popularity_log'], lower_percentile, upper_percentile)
```
