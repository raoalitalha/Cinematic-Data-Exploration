## IQR Value Calculatio

```python
Q1 = df['vote_average'].quantile(0.25)

Q3 = df['vote_average'].quantile(0.75)

IQR = Q3 - Q1

IQR
```

## Removing Outlier from the column (vote_average)

```python
lower_limit = Q1-1.5*IQR

lower_limit
```

```python
upper_limit = Q3+1.5*IQR

upper_limit
```

```python
df[(df['vote_average'] < lower_limit) | (df['vote_average'] > upper_limit)]
```

```python
df_outliers = df[(df['vote_average'] > lower_limit) &  (df['vote_average'] < upper_limit)]

df_outliers
```

```python
sns.boxplot(y= 'vote_average', data=df_outliers)
```
